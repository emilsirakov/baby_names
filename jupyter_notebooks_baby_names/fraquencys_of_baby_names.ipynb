```python
def pars_data(year):
    file_path = (f"../data/names/yob{year}.txt")
    df = pd.read_csv(file_path, names=["names", "gender", "frequency"])
    df["year"] = year
    return df
```

```python
new_df_2000 = pars_data(2000)
x_2000 = new_df_2000
x_2000
```

```python
x = x_2000["frequency"]

bins = 5000,10000,20000, 30000

plt.hist(x, bins, histtype="bar", label= "Frequency", rwidth =0.8)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Frequency')

plt.show();
```

```python
new_df_1950 = pars_data(1950)
new_df_1950
```
