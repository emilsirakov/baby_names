```python
births_dict = {'year':[], 'total_births':[]}
```

```python
def pars_data(year):
    file_path = (f"../data/names/yob{year}.txt")
    df = pd.read_csv(file_path, names=["names", "gender", "frequency"])
    df["year"] = year
    return df
```

```python
pars_data(1990)["frequency"].sum()
```

```python
for year in range(1880, 2022):
    births_dict["year"].append(year)
    total = pars_data(year)["frequency"].sum()
    births_dict["total_births"].append(total)
```

```python
births_dict
```

```python
new_df = pd.DataFrame(births_dict)
new_df
```

```python
yearly_births = new_df
yearly_births
```

```python
x = yearly_births["year"]
y = yearly_births["total_births"]


plt.style.use("ggplot")
plt.figure(figsize=(7.5, 5.8))
plt.bar(x, y, label="Total births/year", color="navy")


plt.title("Total births/year")


plt.xlabel("year")
plt.ylabel("total_births")

plt.legend();
```
